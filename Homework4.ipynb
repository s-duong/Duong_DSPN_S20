$MSE(\beta_0, \beta_1)=E[ (Y- \beta_0 - \beta_1 X)^2]$
$= E[Y^2] - \beta_0 E[Y] - \beta_1 E[XY] - \beta_0 E[Y] + \beta_0^2 + \beta_0\beta_1 E[X] - \beta_1 E[XY] + \beta_0\beta_1 E[X] + (\beta_1)^2 E[X^2]$
$= E[Y^2] - 2 \beta_0 E[Y] - 2 \beta_1 E[XY] + \beta_0^2 + 2 \beta_0\beta_1 E[X] + \beta_1^2 E[X^2]$
$= E[Y^2] - 2 \beta_0 E[Y] - 2 \beta_1 Cov[X,Y] - 2 \beta_1 E[X]E[Y] + \beta_0^2 + 2 \beta_0 \beta_1 E[X] + \beta_1^2 Var[X] + \beta_1^2 E[X]^2$

The slope of the regression line and the pearson correlation coefficient are equal when the standard deviations of the response and predictor variables are equal (the ratio of the standard deviations would be 1). This occurs when the variables are standardized. The correlation coefficient equals the standardized slope of the simple linear regression.

$R^2 = 1 - \frac{RSS}{TSS}$
$= 1 - \frac{\sum_{i=1}^n (y_i - \hat{y_i})^2}{\sum_{i=1}^n (y_i - \bar{y_i})^2}$<br><br>
The coefficient of determination ($R^2$) measures the percentage of variance in our outcome variable that is explained by our predictor variable. This involves estimating the ratio of the residual sums of squares (RSS) to the total sums of squares (TSS). As the RSS decreases relative to the TSS, $R^2$ increases and model fit increases.<br><br>

$RSE= \sqrt{ \frac{RSS}{n-2} } = \sqrt{ \frac{ \sum_{i=1}^{n} (y-\hat{y})^2 }{n-2} }$<br><br>
The residual standard error (RSE) tells us how much of our observed error $\hat{y}$ explains. As the RSS decreases, the RSE decreases, reflecting increasingly better model fit.<br><br>

$F = \frac{ \frac{(TSS-RSS)}{p}} { \frac{RSS}{n - p - 1}} = \frac{1/p\sum_{i=1}^n (y_i - \bar{y_i})^2 - (y_i - \hat{y_i})^2}{1/(n-p-1)\sum_{i=1}^n (y_i - \hat{y_i})^2}$<br><br>
The F-statistic tells us whether the variance that is not explained by our model is greater than the variance explained by our model. As the variances diverge (where TSS > RSS), the ratio increases, reflecting increasingly better model fit.


```r
# -------------------------------
# Question 5(a)
# -------------------------------
setwd('/Users/shirleyduong/Desktop/Duong_DSPN_S20')
d<-read.csv("unrestricted_trimmed_1_7_2020_10_50_44.csv")
invisible(library(dplyr))
invisible(select(d,Subject,Gender,Flanker_Unadj, FS_Tot_WM_Vol,FS_Total_GM_Vol))
```

```r
# -------------------------------
# Question 5(b)
# -------------------------------
lm.fit=lm(Flanker_Unadj~FS_Total_GM_Vol, data=d, na.action = na.exclude)
summary(lm.fit)
coef(lm.fit)
confint(lm.fit)
```

```r
# -------------------------------
# Question 5(c)
# -------------------------------
plot(lm.fit)
#in the residuals vs. fitted values plot, the residuals appear randomly around the 0 line and no residuals stand out from this random pattern
#in the q-qplot, the points fall along the reference line (representing a normal distribution) so we can see that Y is normal
#the scale-location plot shows us how the residuals are spread along the values of predictors; here the residuals are randomly spread 
#in the residuals vs. leverage plot, we see no groups or clusters of extreme values along the leverage dimension 
#yes, it looks like the data is meeting the assumptions of linear regression
```

```r
d$predict <- predict(lm.fit) 
d$resid <- residuals(lm.fit)
library(ggplot2)
ggplot(d, aes(x = FS_Total_GM_Vol, y = Flanker_Unadj)) + geom_smooth(method=lm, se=FALSE) + geom_point(shape=1) + labs(x ="Total GM Volume", y = "Flanker Task Performance")
ggplot(d, aes(x = FS_Total_GM_Vol, y = Flanker_Unadj)) + geom_segment(aes(xend = FS_Total_GM_Vol, yend = predict)) + geom_point(aes(y = predict), shape = 1) + labs(x ="Total GM Volume", y = "Flanker Task Performance")
#the second plot was made to visualize the residuals but it looks noisy...
```

```r
# -------------------------------
# Question 6
# -------------------------------
totGM_values = data.frame(FS_Total_GM_Vol=c(612432, 702562, 596433, 695982, 649230)) 
prediction = data.frame(predict(lm.fit, totGM_values, interval = 'confidence'))  
pred_table = cbind(totGM_values, prediction)
colnames(pred_table) = c('totGM','predicted fit','lower_CI_bound', 'upper_CI_bound')
round(pred_table,2)
```

```r
# -------------------------------
# Question 7(a)
# -------------------------------
lm.multiple=lm(Flanker_Unadj~FS_Total_GM_Vol+FS_Tot_WM_Vol+Gender, data=d)
summary(lm.multiple)
```

```r
# -------------------------------
# Question 7(b)
# -------------------------------
lm.multiple_new=update(lm.multiple, ~.-Gender -FS_Tot_WM_Vol)
summary(lm.multiple_new)
#When we remove the additional predictor variables, the model fit statistics change; in this new model, the RSE is lower, the R-Squared is lower, and the F-stat is greater.
```
