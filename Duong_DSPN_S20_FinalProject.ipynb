**<font size="4">Introduction</font>**

Our basic intuitions of quantity, or number sense, guide many of the everyday choices that we make. We select the line with the fewest people to stand in at the grocery store checkout, the mall food court, or the airport security screening area, without explicitly counting the number of people or items. This ability to rapidly and intuitively estimate and compare quantities larger than 4 is supported by the Approximate Number System (ANS), which is present in infancy (e.g., Libertus & Brannon, 2009) and in non-human animals, such as primates, birds, and fish (e.g., Vallortigara, 2017). The ANS is typically measured with a numerosity comparison task, in which participants are presented with two arrays of dots and are instructed to indicate which array is more numerous (Figure 1). Performance is ratio-dependent, such that numerosities that are closer to together are more difficult to discriminate (e.g., 12 vs. 24 is easier than 20 vs. 24; Libertus & Brannon, 2009). Individual differences in ANS acuity are associated with children’s and adults’ formal math abilities (e.g., Halberda et al., 2008; Libertus et al., 2013; Guillaume et al., 2013) and may reflect differences in cognitive processes that are involved in lab based ANS tasks, such as vision or attention. Such processes can be approximated by the eye movement or tracking strategies that people employ during these tasks. 

Researchers have used eye-tracking (ET) measures, such the proportion of time spent looking at an area of interest and the number of switches between areas, to make inferences about the association between looking and thinking (see Lai et al., 2013 for a review; also see Orquin & Holmqvist, 2019 for a review of potential issues in ET research). Although much of this work is dominated by studies about reading and language processing (Aryadoust & Ang, 2019), research in the domain of numerical cognition is emerging (Hartmann & Fischer, 2016 for a review). Few studies have examined the ET patterns of participants while completing a dot comparison task, showing that participants’ eye movements are associated with the overall quantity of dots within arrays (Cheyette & Piantadosi, 2019) and task features, such as difficulty (i.e., the ratio of the numerosities; Odic & Halberda, 2019). What remains understudied are how ET patterns relate to individual variability in ANS precision and the role of formal math skills in this association. For instance, there may be similarities between how individuals process number at a basic level and how they process numerical information on a standardized test. It is possible that individuals with higher math skills are more likely to appropriately attend to relevant information in both tasks and thus, respond more accurately. Understanding these interrelations may provide insight into the mechanisms that underlie the link between the ANS, a foundational cognitive system, and formal math abilities.

The present study examines ANS precision, eye movement patterns, and math achievement in adults, with a focus on whether the association between eye movements and ANS acuity is moderated by math ability. The specific ET variable is the frequency with which individuals switch between arrays in a dot comparison task, a method that people may use to build their representation of quantity before making a judgement.

**<font size="4">Method</font>**

The data were derived from a larger study examining the ANS and related cognitive functions (see Braham & Libertus, 2018 and Braham et al., 2019 for papers using this dataset). Adult college students’ (final sample: *n*=158) ANS acuity and math achievement were measured with a dot comparison task (Figure 1) and the Woodcock Johnson Applied Problems subtest (Woodcock et al., 2001), respectively. In the ANS task, participants were simultaneously shown arrays of yellow and blue dots on a gray background and asked to indicate which array contained more dots. Each array of dots contained between 12 and 36 dots of various sizes, and participants were shown 90 pairs of arrays (trials). Each trial began with a fixation cross that appeared for 500 ms, followed by trials that appeared for 1500 ms. Trial difficulty was determined by the ratio between the larger to smaller numbers of dots; there were five ratio categories, with 18 trials per group:  1.11, 1.14, 1.2, 1.25, and 1.33. The total surface area of each array was varied to control for participants’ potential reliance on other perceptual cues. The dependent measures are accuracy, response time (RT), and the frequency with which participants switched between arrays. Summary statistics of participants’ performance in both tasks show considerable individual variability (Table 1).

**<font size="4">Variables of Interest</font>**

1. **Subject number** (ID): Note that ID numbers take the form of "p1/p2/p3/p4_[##]" where p1 to p4 are arbitrary labels representing the period or semesters during which the data was collected, e.g., some participants were tested in spring 2014 and others in fall 2014. This distinction is not relevant to the purposes of this study as data from all time periods were included in the analyses.

*Main variables*
2. **ANS task accuracy** (ans_acc): The percentage of correct trials in the dot comparison task. Scores range from 0 to 1.
3. **Switches between arrays in the ANS task** (ans_switch): The number of times participants switched between sets of dots in each trial. This was calculated by summing the number of times three or more gaze points occurred on one array after gaze points on the other side occurred. An average number of switches for each ratio category was calculated. 
4. **WJ Applied Problems scores** (applied_problems): The measure of math ability; the standardized score is the dependent measure.

*Control variables*
5. **Ratio** (ans_ratio): Ratio categories, ranging from 1.11 to 1.33, representing trial difficulty in the ANS task; larger ratios correspond to easier trials. The inclusion of this parameter accounts for the possibility that participants will make more switches in more difficult trials (as shown in Odic & Halberda, 2019). 
6. **Response Time** (ans_rt): Response time (RT) in seconds. The inclusion of this variable accounts for a potential confound, as the number of switches may increase as RT increases.

The continuous independent variables are mean-centered to reduce multicollinearity issues.

**<font size="4">Research Question & Hypothesis</font>**

Does math ability moderate the relation between ANS acuity and the frequency of switches between arrays in a numerosity comparison task?

$$ H_1: Y_{ANSacuity} = \beta_0 + \beta_1 X_{switches} + \beta_2 X_{math} + \beta_3 X_{switches} X_{math} + \beta_4 X_{ratio} + \beta_5 X_{RT}$$ 

It is hypothesized that the relation between switch frequency and ANS precision is modulated by math skills, even after controlling for task difficulty (i.e., ratio) and RT.

$$ H_0: Y_{ANSacuity} = \beta_0 + \beta_1 X_{ratio} + \beta_2 X_{RT}$$

In contrast, there may be no relation between the variables of interest, but ratio and RT would still be linked to ANS acuity as shown in prior work (see Libertus & Brannon, 2009 for a review).

**<font size="4">Data Organization</font>**

Data Cleansing Steps:
1. Select only the data that are relevant to my questions (i.e., the variables of interest above).
2. Remove participants with missing data, poor calibration, failure to complete tasks, other issues noted in the participant run sheets (e.g., experimenter or computer error), or any combination of these criterion (*n*=20). The original sample included 178 individuals and final sample includes 158.
3. Remove outliers based on RT. If a participants’ trial-level RT was 2.5 standard deviations above or below their average RT, the corresponding trials were discarded and missing values were not interpolated. This is consistent with the anomaly removal method (Van Selst & Jolicoeur, 1994) used in previous work with the same dataset (e.g., Braham et al., 2018).

**<font size="4">Generalized Linear Model for Inference</font>**

This analysis tested whether there is a relationship between the number of switches in the comparison task and ANS acuity, and whether this association changes as a function of formal math ability.

```r
#mean-center the continuous predictor variables 

d1<-d #d1 will contain the centered values, so that d can be used for the plot

center_scale <- function(x) {
  scale(x, scale = FALSE)
}

d1$ans_rt<-center_scale(d1$ans_rt)
d1$ans_switch<-center_scale(d1$ans_switch)
d1$applied_problems<-center_scale(d1$applied_problems)

d1$ans_ratio<-as.factor(d1$ans_ratio) #change from numeric to grouping variable 

lm.mod=lm(ans_acc~ans_switch*applied_problems+ans_ratio+ans_rt, data=d1)
summary(lm.mod)
```

The generalized linear model shows that there are positive associations between the number of switches between arrays and accuracy in the dot comparison task, and standardized math scores and accuracy. Also, math ability moderates the relation between switches and task accuracy, such that magnitude of the association is stronger for individuals with higher math scores.

**Residual plots**

```r
par(mfrow = c(2,2))
plot(lm.mod)
```

The residual plots show that the data meet the assumptions of linear regression.

```r
par(mfrow=c(1,1))

#create factor variables to visualize interaction
d <- d%>%mutate(ap_group = ifelse(applied_problems > mean(applied_problems) + sd(applied_problems), "High",
                                         ifelse(applied_problems < mean(applied_problems) + sd(applied_problems) & 
                                                  applied_problems > mean(applied_problems) - sd(applied_problems), "Average",
                                                "Low")))

library(ggplot2)
library(ggsci)
theme_set(theme_classic())
p=ggplot(d,aes(ans_switch, ans_acc, color=ap_group))+
  geom_smooth(method=lm)+geom_point()+
  labs(x="Number of Switches", y="Accuracy", color="Math Ability")
p + scale_color_jco() + theme(text = element_text(size = 18))
```

The plot shows that the number of switches between dot arrays is related to accuracy in the ANS task and that this association varies as a function of formal math ability. Post-hoc contrast analyses (which were not conducted here) can offer additional, nuanced insight into group differences.

**<font size="4">Bayes Factor Analysis</font>**

This analysis was used quantify the evidence for the alternative hypothesis relative to the null.

```r
library(BayesFactor);
full<-lmBF(ans_acc~ans_switch*applied_problems+ans_ratio+ans_rt, data=d)
null<-lmBF(ans_acc~ans_ratio+ans_rt, data=d)

baes <- c(full, null)
baes

full/null
```

There is strong evidence for the hypothesis that there is an association between the number of switches and ANS acuity, as moderated by math ability, relative to the null hypothesis.

**<font size="4">Cross Validation for the GLM</font>**

Leave one out cross validation (LOOCV) was used to assess the predictive performance of the linear model and evaluate its generalizability. 

NOTE: LOOCV was chosen over k-fold CV due to sample size (i.e., there was worry that there was not enough power in each fold or group). However, since k-fold CV is preferred over LOOCV, supplementary analyses (below) were run with k-fold CV where k=5 and k=10, producing similar results.

```r
library(boot)
lm.mod=glm(ans_acc~ans_switch*applied_problems+ans_ratio+ans_rt, data=d1)
cv.mod=cv.glm(d1, lm.mod) 
print(paste0("Raw LOOCV estimate of prediction error, k=n: ", cv.mod$delta[1]))

#supplementary analyses
cv.mod.5 = cv.glm(d1, lm.mod, K=5) 
print(paste0("CV prediction error, k=5: ", cv.mod.5$delta[1]))

cv.mod.10 = cv.glm(d1, lm.mod, K=10) 
print(paste0("CV prediction error, k=10: ", cv.mod.10$delta[1]))
```

The test error derived from the linear model using LOOCV is approximately .01, suggesting that the model is generalizable to a high degree.

**<font size="4">Conclusion</font>**

The present study tested the hypothesis that there is a relation between eye movements in a numerosity comparison task, specifically the number of switches between arrays, and ANS precision, and that this relation is moderated by math ability. A linear regression analysis revealed an association between these variables, even after accounting for RT and task difficulty. Switching between the dot arrays generally mattered more for individuals with higher math skills and a visualization of the data suggests that task performance did not increase with switches for individuals who scored below-average on the formal math assessment. 

Although the causal mechanisms underlying these findings cannot be derived from this study, there are several possible explanations worth speculating. there are  For instance, switching between the dot arrays is one way that participants can obtain information about the quantity of dots before making their judgement. Individuals with higher math ability may benefit from switching because they are attending to unique parts of the stimulus display, while those with lower math ability are looking to the same or similar areas when switching. Thus, those with higher math skills are building up a more accurate representation of the numerosities, which guides their decision-making. Another potential explanation is that the eye movements of individuals' with higher ANS acuity and formal math ability reflect a domain-general tendency to appropriately attend to relevant task-specific information.  However, participants' eye movement patterns within the arrays between switches may be more informative to our understanding of how looking relates to numerosity processing. It is possible that individuals are attending to particular features of the stimuli, e.g., the more numerous parts of each array, and employ a perceptual subsetting method to build their quantity represenation (as participants do in a dot enumeration task; Paul et al., 2017). Future work should explore a variety of ET patterns to gain a more nuanced understanding of how individuals are looking at (and thus, potentially processing) numerosities, as well as identify what types of eye movements are more predictive of ANS task performance than others.
