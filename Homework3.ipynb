```r
#b) Exp 2: 1,000 independent throws
n_b = 1000
dat_b = data.frame(length=ceiling(runif(n_b, min=x, max=a)))
ggplot(dat_b, aes(length)) + geom_histogram(alpha = 0.9)
```

```r
#c) Exp 3: 10,000 independent throws
n_c = 10000
dat_c = data.frame(length=ceiling(runif(n_c, min=x, max=a)))
ggplot(dat_c, aes(length)) + geom_histogram(alpha = 0.9)
```

```r
# -------------------------------
# Question 3 Response
# -------------------------------
#As the number of dice rolls approaches infinity, the simulated values approach the expected values of a fair roll in which the probability of getting any value is 1 in 6 and are equally likely.


# -------------------------------
# Question 4
# -------------------------------
#a) Exp 1: 10,000 throws, 1 die
n = 10000
norm_dat=data.frame(length=rnorm(n, mean=0, sd=1))
dat_1 = data.frame(length=ceiling(runif(n, min=x, max=a)))
ggplot(dat_1, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat_1, aes(sample=length)) + stat_qq(distribution = qnorm) + ggtitle("Q-Q Plot: 1 Die vs. Normal Distribution") 
```

```r
#b) Exp 2: 10,000 throws, 2 dice
mutate(dat_1, dat_2=sum(dat_1+dat_1))
ggplot(dat_2, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat_2, aes(sample=length)) + stat_qq(distribution = qnorm) + ggtitle("Q-Q Plot: 2 Die vs. Normal Distribution") 
```

```r
#c) Exp 3: 10,000 throws, 3 dice
mutate(dat_2, dat_3=sum(dat_1+dat_2))
ggplot(dat_3, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat_3, aes(sample=length)) + stat_qq(distribution = qnorm) + ggtitle("Q-Q Plot: 3 Die vs. Normal Distribution") 
```

```r
#d) Exp 4; 10,000 throws, 6 dice
mutate(dat_3, dat_6=sum(dat_3+dat_3))
ggplot(dat_6, aes(length)) + geom_histogram(alpha = 0.9)
ggplot(dat_6, aes(sample=length)) + stat_qq(distribution = qnorm) + ggtitle("Q-Q Plot: 6 Die vs. Normal Distribution") 
```
